## dessiner une croix

Soit un entier positif impair n.
affichez une croix au sein d'un carré de coté n.

Par exemple, pour n = 5 vous obtiendrez :

    o       o    
      o   o      
        o         
      o   o
    o       o




```python
# Votre code ICI
n = 9
nbBlancs = n-2
for i in range(n//2):
    for j in range(i):
        print(" ",end = "")
    print("o",end = "")
    for j in range(nbBlancs):
        print(" ",end = "")
    print("o")
    nbBlancs = nbBlancs-2
for i in range(n//2):
    print(" ",end = "")
print("o")
nbBlancs = 1
for i in range(n//2):
    for j in range(n//2-1-i):
        print(" ",end = "")
    print("o",end = "")
    for j in range(nbBlancs):
        print(" ",end = "")
    print("o")
    nbBlancs = nbBlancs+2
        
```

## Tables de multiplication

Soit un enrier ninférieur ou égal à 10.
écrire un programme qui affiche les tables de multiplication de tous les entiers entre 1 et n.
Par exemple avec n = 6, vous afficherez avec un alignement vertical correct :



```python
# Votre code ICI
n = 10
for i in range(n):
    for j in range(10):
        r = (i+1)*(j+1)
        if (r < 10):
            print("   ",r,end = "")
        else:
            if (r<100):
                print("  ",r,end = "")
            else:
                print(" ",r)
    print("")
```

## Triangle de Pascal

1) Réaliser un programme qui affiche les n (n >=2) premières lignes du triangle de Pascal.
Pour n = 10, on obtient :



l'élément situé en ligne i et colonne j, noté $t_{i,j}$ vaut la somme de celui situé en ligne i-1,colonne j avec celui situé en ligne i-1 et colonne j-1. 

cad : $t_{i,j}$ = $t_{i-1,j}$ + $t_{i-1,j-1}$

```python
# Votre code ICI
n = 10
l = [[1],[1,1]]
for i in range(2,n+1):
    l.append([])
    l[i].append(1)
    for j in range(i-1,0,-1):
        l[i].append(l[i-1][j]+l[i-1][j-1])
    l[i].append(1)
    
for i in range(n):
    print(l[i])
    
```

## Rendre la monnaie

1) Écrire un programme qui affiche toutes les manières possibles d'obtenir 1 Euro
avec des pièces de 2 centimes, 5 centimes et 10 centimes. 
On veillera à ne pas afficher les 0 : 20$\times$5 = 100 et non 0$\times$2+20$\times$5$\times$0$\times$10=100. 
Le programme calculera et affichera le nombre total de manières
d’obtenir ainsi 1 Euro... On utilisera 3 boucles `for` imbriquées.


```python
# Votre code ICI
for i in range(11):
    for j in range(21):
        for k in range(51):
            if ( i*10 + j*5 + k*2 == 100 ) :
                res = []
                if (i != 0 ):
                    res.append(str(i)+ "x10") 
                if (j != 0 ):
                    res.append(str(j)+ "x5")
                if (k != 0 ):
                    res.append(str(k) + "x2")
                A = res[0]
                for i in range(len(res)-1):
                    A = A+" + "
                if (len(res)>= 2):
                    A = A+ res[len(res)-1] + " = 100"
                else :
                    A = A+ " = 100"
                print(A)
```

## Produit matrice - vecteur

Faire un programme réalisant un produit matrice vecteur. La matrice est stockée dans une liste de liste, le vecteur dans une liste.

On rappele la formule avec M = ($M_{i,j}$) i = 0,..,n-1 indice des lignes et j = 0,..,p-1 indice des colonnes, et V = ($V_j$) j = 0,..,p-1 :

Mv = w, avec $w_i$ = $\sum_{j = 0}^{p-1}A_{i,j}v_j$. i = 0,..,n-1.

Voici un exemple :



```python
# Votre code ICI
M=[[1,2,3],[4,5,6],[7,8,9]]
V = [1,-2,0.5]
R = V[:]
if (len(V) == len(M[0])):
    for i in range(len(R)):
        R[i] = 0
        for j in range(len(M[0])):
            R[i] = R[i]+M[i][j]*V[j]
print(R)
                   
```

## Produit Matrice - Matrice

Soit une matrice M à n lignes, p colonnes stockée dans une liste de n listes à p éléments.
Soit une matrice Q à p lignes et k colonnes stockée dans une liste de p listes à k éléments.

Faites un programme réalisant le produit de M par Q. Cela donne une matrice R ayant n lignes et k colonnes.

La formule est :

$R_{i,l}$ = $\sum_{j = 0}^{p-1}A_{i,j}Q_{j,l}$. i = 0,..,n-1, l = 0,..,k-1.

voici un exemple de calcul :


```python
# Votre code ICI
M = [[1,2,3],[4,5,6]]
Q = [[7,8],[9,-1],[-2,-3]]

R = []

if(len(M[0]) == len(Q)):

    for i in range(len(M)):
        R.append([])
        for j in range(len(Q[0])):
            R[i].append(0)

    for i in range(len(M)):
        for j in range(len(Q[0])):
            for k in range(len(M[0])):
                R[i][j] = R[i][j] + M[i][k]*Q[k][j]

print(R)
```
